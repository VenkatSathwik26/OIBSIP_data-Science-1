```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('iris.csv.zip')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Species'].value_counts().plot(kind='pie',colors=['blue','green','red'],autopct='%2f')
```

```python
df.duplicated().sum()
```

```python
X=df.drop(['Species'],axis=1).values
Y=df['Species'].values
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['Species']=le.fit_transform(df['Species'])
```

```python
df.sample(5)
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(random_state=42)
```

```python
rfc.fit(X_train,y_train)
```

```python
y_pred=rfc.predict(X_test)
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import classification_report,accuracy_score
accuracy=accuracy_score(y_test,y_pred)
print(f'Accuracy:{accuracy}')
```

```python
report=classification_report(y_test,y_pred)
print(f'classification Report:\n{report}')
```

```python
importances=rfc.feature_importances_
feature_names=df.columns[:-1]
plt.figure(figsize=(10,6))
plt.barh(range(len(importances)),importances,align='center')
plt.yticks(range(len(importances)),feature_names)
plt.xlabel('Feature Importance')
plt.title('Feature Importance Plot')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns 
import matplotlib.pyplot as plt

cm=confusion_matrix(y_test,y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm,annot=True,fmt="d",cmap="Blues")
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
import numpy as np
y_setosa_binary=np.where(y_test=='Iris-setosa',1,0)
y_pred_setosa_proba=rfc.predict_proba(X_test)[:,0]

fpr,tpr,_=roc_curve(y_setosa_binary,y_pred_setosa_proba)
auc=roc_auc_score(y_setosa_binary,y_pred_setosa_proba)

plt.figure(figsize=(8,6))
plt.plot(fpr,tpr,label=f'ROC Curve(AUC={auc:2f})')
plt.plot([0,1],[0,1],'k--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```
